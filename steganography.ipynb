```python
from PIL import Image
from cryptography.fernet import Fernet
import base64

# ---------- Encryption Helpers ----------

def generate_key():
    return Fernet.generate_key()

def encrypt_message(message, key):
    f = Fernet(key)
    return f.encrypt(message.encode()).decode()

def decrypt_message(encrypted_message, key):
    f = Fernet(key)
    return f.decrypt(encrypted_message.encode()).decode()

# ---------- Steganography Helpers ----------

def text_to_binary(message):
    return ''.join(format(ord(char), '08b') for char in message)

def binary_to_text(binary):
    chars = [binary[i:i+8] for i in range(0, len(binary), 8)]
    return ''.join([chr(int(char, 2)) for char in chars])

def encode_image(input_image_path, output_image_path, encrypted_message):
    image = Image.open(input_image_path)
    if image.mode != 'RGB':
        image = image.convert('RGB')
    encoded = image.copy()
    width, height = image.size

    message_binary = text_to_binary(encrypted_message) + '1111111111111110'  # EOF marker
    data_index = 0

    for y in range(height):
        for x in range(width):
            pixel = list(image.getpixel((x, y)))
            for i in range(3):  # RGB channels
                if data_index < len(message_binary):
                    pixel[i] = pixel[i] & ~1 | int(message_binary[data_index])
                    data_index += 1
            encoded.putpixel((x, y), tuple(pixel))
            if data_index >= len(message_binary):
                break
        if data_index >= len(message_binary):
            break

    encoded.save(output_image_path)
    print(f"[+] Message encoded and saved to {output_image_path}")

def decode_image(stego_image_path):
    image = Image.open(stego_image_path)
    binary_data = ''
    for y in range(image.height):
        for x in range(image.width):
            pixel = image.getpixel((x, y))
            for i in range(3):
                binary_data += str(pixel[i] & 1)

    eof_index = binary_data.find('1111111111111110')
    if eof_index != -1:
        message_binary = binary_data[:eof_index]
        return binary_to_text(message_binary)
    else:
        return "[!] No hidden message found"

# ---------- Main Program ----------

if __name__ == '__main__':
    choice = input("Do you want to (E)ncode or (D)ecode? ").upper()

    if choice == 'E':
        original = input("Enter path to original image (e.g., input.png): ")
        output = input("Enter path to save encoded image (e.g., output.png): ")
        secret = input("Enter the secret message: ")

        key = generate_key()
        encrypted = encrypt_message(secret, key)
        print(f"[+] Encryption Key (SAVE THIS!): {key.decode()}")

        encode_image(original, output, encrypted)

    elif choice == 'D':
        stego = input("Enter path to stego image: ")
        key = input("Enter the secret key: ").encode()

        extracted = decode_image(stego)
        try:
            decrypted = decrypt_message(extracted, key)
            print(f"[+] Hidden Message: {decrypted}")
        except Exception as e:
            print(f"[!] Decryption failed: {e}")
    else:
        print("Invalid choice. Please choose E or D.")
```
